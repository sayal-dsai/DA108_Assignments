```python
import cv2
import matplotlib.pyplot as plt
```

```python
original_image = cv2.imread("Orb.jpeg")
# plt.imshow(original_image)
rgb_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
greyscale_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
# plt.imshow(greyscale_image)
_, bw_image = cv2.threshold(greyscale_image, 127, 255, cv2.THRESH_BINARY)
# plt.imshow(bw_image)
```

```python
plt.figure()

plt.subplot(1,3,1)
plt.imshow(rgb_image)
plt.title("Origial colour image")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(greyscale_image, cmap='gray')
plt.title("Grey image")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(bw_image, cmap='gray')
plt.title("BW image")
plt.axis('off')

plt.show()
```

```python
cv2.imwrite("Grey_Orb.jpeg", greyscale_image)
cv2.imwrite("BW Orb.jpeg", bw_image)
```
